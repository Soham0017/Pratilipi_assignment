```python
# Pratilipi publications over time
metadata['Month_published'] = metadata['published_at'].dt.to_period('M')
metadata['Month_published'].value_counts().sort_index().plot(kind='line', figsize=(14, 6), marker='o')
plt.title('Pratilipi Publications Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Publications')
plt.show()
```

```python
metadata['Month_updated'] = metadata['updated_at'].dt.to_period('M')
metadata['Month_updated'].value_counts().sort_index().plot(kind='line', figsize=(14, 6), marker='o')
plt.title('Pratilipi Last Updations Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Updations')
plt.show()
```

```python
# Most published categories
plt.figure(figsize=(12, 8))
category_counts = metadata['category_name'].value_counts()
sns.barplot( x=category_counts.values, y=category_counts.index, palette = 'viridis')
plt.title('Most Published Categories')
plt.xlabel('Number of Pratilipis (log scale)')
plt.ylabel('Category')
plt.xscale("log")
plt.grid(axis='x', linestyle='--', alpha=0.7)  # Add grid lines
plt.xlim(100, 1000000) # Ensures log scale between 10^2 to 10^6
plt.show()
```

```python
# Most Popular category according to user interaction
merged_df = user_interaction.merge(metadata, on="pratilipi_id", how="left")

plt.figure(figsize=(12, 8))
category_counts = merged_df['category_name'].value_counts()
sns.barplot(x=category_counts.values, y=category_counts.index, palette = 'viridis')
plt.title('Most Popular Categories Based on User Interactions')
plt.xlabel('Number of User Interactions (log scale)')
plt.ylabel('Category')
plt.xscale("log")
plt.xlim(100, 10000000)
plt.grid(axis='x', linestyle='--', alpha=0.7)  # Add grid lines
plt.show()
```

```python
# mean reading time of each category according to metadata
plt.figure(figsize=(12, 8))
mean_reading_time = metadata.groupby('category_name')['reading_time'].mean().sort_values()
sns.barplot( x=mean_reading_time.values, y=mean_reading_time.index, palette = 'coolwarm')
plt.title('Mean Reading Time for Each Pratilipi Category')
plt.xlabel('Mean Reading Time (minutes)')
plt.ylabel('Category')
plt.grid(axis='x', linestyle='--', alpha=0.7)  # Add grid lines
plt.show()
```

```python
print('Total user interactions :', len(user_interaction))
print('Total unique users :', user_interaction['user_id'].nunique())
print('Interactions per user :', len(user_interaction) / user_interaction['user_id'].nunique())
```

```python
def get_user_reading_info(user_id):

    # Returns the unique categories a user has interacted with, their total interaction count, unique authors.
    user_data = merged_df[merged_df['user_id'] == user_id]
    categories = user_data['category_name'].unique().tolist()
    authors = user_data['category_name'].unique().tolist()
    interaction_count = user_data.shape[0]
    return categories, interaction_count, len(authors)

# Example 
user_id = 5506791978962946 
categories_read, total_interactions, total_authors = get_user_reading_info(user_id)

print(f"User {user_id} has interacted {total_interactions} times.")
print(f"Categories read by User {user_id}: {categories_read}")
print(f"Unique Authors read by User : {total_authors}")
```

# Feature Engineering

```python
# total users for each pratilipi
pratilipi = (
    user_interaction.groupby('pratilipi_id')['user_id']
    .count()
    .rename('total_users')
    .to_frame()
    .sort_values(by='total_users', ascending=False)
    .reset_index()
)

# avg_read_percent of each pratilipi
avg_read = (
    user_interaction.groupby('pratilipi_id')['read_percent']
    .mean()
    .rename('avg_read_percent')
    .to_frame()
    .sort_values(by='avg_read_percent', ascending=False)
    .reset_index()
)

pratilipi = pratilipi.merge(avg_read, left_on = 'pratilipi_id', right_on = 'pratilipi_id')

metadata = metadata.merge(pratilipi, left_on = 'pratilipi_id', right_on = 'pratilipi_id')
metadata.head()
```
